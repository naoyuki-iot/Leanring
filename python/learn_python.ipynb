```python
tokyo_temps=[15.1,15.4,15.2] #list
plt.plot(tokyo_temps)
```

```python
tokyo_sum=sum(tokyo_temps)
tokyo_sum
```

```python
tokyo_max=max(tokyo_temps)
tokyo_max
```

```python
tokyo_min=min(tokyo_temps)
tokyo_min
```

```python
tokyo_len=len(tokyo_temps)
tokyo_len
```

```python
tokyo_average=tokyo_sum/tokyo_len
tokyo_average
```

```python
plt.bar([0,1,2],tokyo_temps)
plt.plot([0,len(tokyo_temps)],[tokyo_average,tokyo_average],color="red")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np 
import math
plt.figure(figsize=(4,4), dpi=100) 
plt.title('Quarter Unit Circle') 
plt.grid() 
x,y=[],[]

for theta in np.linspace(0,90):
    x.append(math.sin(math.radians(theta)))
    y.append(math.cos(math.radians(theta)))

plt.plot(x,y)
plt.show()
```
