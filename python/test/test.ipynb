```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np 
import math
plt.figure(figsize=(4,4), dpi=100) 
plt.title('Quarter Unit Circle') 
plt.grid() 
x,y=[],[]

for theta in np.linspace(0,90):
    x.append(math.sin(math.radians(theta)))
    y.append(math.cos(math.radians(theta)))

plt.plot(x,y)
plt.show()
```
